```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re
import string
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import confusion_matrix
```

```python
df.columns
```

```python
df.drop(["title", "subject", "date"], axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
 df =df.sample(frac = 1)
```

```python
df.reset_index(inplace = True)
df.drop(["index"], axis = 1, inplace = True)
```

```python
df.head(15)
```

```python
def word_drop(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text
```

```python
df["text"] = df["text"].apply(word_drop)
```

```python
df.head(15)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize= (20,20))
wc = WordCloud(max_words = 2000, width = 1600, height = 800, stopwords = STOPWORDS).generate(" ".join(df[df.catagory == 0].text))
plt.imshow(wc, interpolation = 'bilinear')
```

```python
plt.figure(figsize= (20,20))
wc = WordCloud(max_words = 2000, width = 1600, height = 800, stopwords = STOPWORDS).generate(" ".join(df[df.catagory == 1].text))
plt.imshow(wc, interpolation = 'bilinear')
```

```python
x = df["text"]
y = df["catagory"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
Predict_LR = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, Predict_LR))
```

```python
cf_matrix = confusion_matrix(y_test, Predict_LR)

print(cf_matrix)
```

```python
ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
